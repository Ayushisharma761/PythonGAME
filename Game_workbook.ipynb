# 5.Find the 95%, 90%, and 99%, confidence interval for 'Potential','Wage','Weight' column.

```python
# Remove 'lbs' from the 'Weight' column and convert it to float
game_df['Weight'] = game_df['Weight'].str.replace('lbs', '').astype(float)
```

```python
import pandas as pd
import numpy as np
from scipy.stats import norm

# Define a function to calculate confidence intervals
def calculate_confidence_interval(data, confidence_level):
    n = len(data)
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # Using sample standard deviation (ddof=1)
    z_critical = norm.ppf(1 - (1 - confidence_level) / 2)
    margin_of_error = z_critical * (std / np.sqrt(n))
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error
    return lower_bound, upper_bound

# Columns for which to calculate confidence intervals
columns_to_analyze = ['Potential', 'Wage', 'Weight']

# Calculate confidence intervals for each column
confidence_levels = [0.95, 0.90, 0.99]
confidence_intervals = {}

for column in columns_to_analyze:
    data = game_df[column].dropna()  # Remove missing values
    confidence_intervals[column] = {}
    for confidence_level in confidence_levels:
        lower_bound, upper_bound = calculate_confidence_interval(data, confidence_level)
        confidence_intervals[column][confidence_level] = (lower_bound, upper_bound)

# Print the confidence intervals
for column in columns_to_analyze:
    print(f"Confidence Intervals for '{column}' column:")
    for confidence_level in confidence_levels:
        lower_bound, upper_bound = confidence_intervals[column][confidence_level]
        print(f"{confidence_level * 100}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
    print()
```

The confidence intervals provide a range of values within which we can estimate the true population parameter (mean) with a certain level of confidence. In this context:

1. Confidence Intervals for 'Potential' column:
   - 95.0% Confidence Interval: (71.22, 71.40)
   - 90.0% Confidence Interval: (71.23, 71.38)
   - 99.0% Confidence Interval: (71.19, 71.42)

   These intervals indicate that with 95% confidence, the true mean of the 'Potential' column is estimated to lie between 71.22 and 71.40. Similarly, with 90% confidence, the mean is estimated to lie between 71.23 and 71.38, and with 99% confidence, it is estimated to lie between 71.19 and 71.42.

# 6. Prove Central Limit Theorom by using 'potential' column of the game_data.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm

# Sample size and number of samples
sample_size = 30
num_samples = 1000

# Function to calculate the mean of a sample
def sample_mean(sample_data):
    return np.mean(sample_data)

# Generating sample means
sample_means = []
for _ in range(num_samples):
    sample = game_df['Potential'].sample(n=sample_size, replace=True)
    sample_means.append(sample_mean(sample))

# Population mean and standard deviation
population_mean = game_df['Potential'].mean()
population_std = game_df['Potential'].std()

# Sample mean and standard deviation
sample_mean_mean = np.mean(sample_means)
sample_mean_std = np.std(sample_means)

# Plotting the distribution of sample means
plt.figure(figsize=(8, 6))
plt.hist(sample_means, bins=30, color='purple', edgecolor='black', alpha=0.7, density=True, label='Sample Means')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.title('Distribution of Sample Means (Central Limit Theorem)')

# Plotting the normal distribution curve using population mean and std
x = np.linspace(min(sample_means), max(sample_means), 100)
plt.plot(x, norm.pdf(x, population_mean, population_std), color='blue', linewidth=2, label='Population Distribution')

plt.legend()
plt.show()

# Print statistics
print("Population Mean:", population_mean)
print("Population Standard Deviation:", population_std)
print("Sample Mean of Sample Means:", sample_mean_mean)
print("Sample Standard Deviation of Sample Means:", sample_mean_std)
```

The Central Limit Theorem (CLT) states that as the sample size increases, the distribution of sample means will tend to follow a normal distribution, regardless of the shape of the original population distribution, as long as the sample size is large enough.

A random sample of size 30 from the 'Potential' column is considered and repeated this process 1000 times to create 1000 sample means. Then plotted a histogram of these sample means and compared it with the theoretical normal distribution of the population.

The results show that the population mean (71.31) and the sample mean of sample means (71.29) are very close to each other. Similarly, the population standard deviation (6.14) and the sample standard deviation of sample means (1.12) are also relatively close. This aligns with the Central Limit Theorem's prediction that as the sample size increases, the sample means will become more and more normally distributed, and their mean and standard deviation will approximate the population mean and standard deviation, respectively.

The histogram plot of the sample means also shows a bell-shaped curve, resembling a normal distribution. The blue curve represents the theoretical normal distribution based on the population mean and standard deviation. The more sample means we take, the closer the histogram will resemble the normal distribution curve.
